```python
df = pd.DataFrame(vetor_resultados, columns=['Accuracy', 'Recall', 'Precision', 'Kappa'])

# Crie uma figura e eixos para o gráfico
fig, ax = plt.subplots(figsize=(10, 6))

# Crie um boxplot para as colunas do DataFrame
sns.boxplot(data=df[['Accuracy', 'Recall', 'Precision', 'Kappa']], ax=ax, palette='Set1', width=0.5)

# Crie um stripplot para as mesmas colunas (usando um eixo diferente)
sns.stripplot(data=df[['Accuracy', 'Recall', 'Precision', 'Kappa']], ax=ax, jitter=True, color='black', alpha=0.5)

# Rotule o eixo y
ax.set_ylabel('Valor')

# Defina o título do gráfico
ax.set_title('Boxplot e Stripplot das Métricas')

# Exiba o gráfico
plt.show()
```
